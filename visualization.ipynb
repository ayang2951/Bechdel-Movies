```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
bechdel = pd.read_csv('data_ingest/bechdel/bechdel_data.csv')
imdb = pd.read_csv('data_ingest/imdb/imdb_data.csv')
```

```python
bechdel = bechdel.rename(columns = {'imdb_id': 'imdbID'})
```

```python
data = bechdel.merge(imdb, on = 'imdbID', how = 'inner')
```

```python
rating_over_years = data[['year', 'rating']].groupby(by = 'year').mean()
```

```python
plt.plot(rating_over_years.index, rating_over_years['rating'])
plt.show()
```

```python
plt.scatter(data['imdbVotes'], data['rating'], alpha = 0.1)
plt.show()
```

```python
plt.scatter(data['year'], data['rating'], alpha = 0.1)
plt.show()
```

```python
plt.scatter(data['year'], data['imdbVotes'], alpha = 0.05)
plt.show()
```

```python
unique_titles = data.drop_duplicates(['title'])
unique_titles.sort_values(by = ['imdbVotes'], axis = 0, ascending = False).head(50)
```
